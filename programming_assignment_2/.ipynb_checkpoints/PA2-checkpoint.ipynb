```python
import math
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
from pandas.api.types import is_numeric_dtype
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import SGDClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.decomposition import PCA
from sklearn.metrics import mean_squared_error

%matplotlib inline
```

```python
#Part 1

# Several columns in this dataset have missing data. Before we can use ML algorithms on this data, 
# we need to impute these missing values. Replace all of the numeric missing values (NaNâ€™s) with 
# the mean of their respective columns. Identify which columns had missing values and what the mean 
# was that you used to replace them.
```

```python
train = pd.read_csv("train.csv", header=0, sep=',')
# train.set_index("Id")

train_column_averages = {}

for column in train:
    temp_list = train[column].tolist()
    if is_numeric_dtype(train[column]):
        temp_avg = 0
        for value in temp_list:
            if not np.isnan(value):
                temp_avg += value
        train_column_averages[column] = temp_avg/len(temp_list)
    else:
        temp_avg = {}
        for value in temp_list:
            if not pd.isnull(value):
                if temp_avg.get(value) == None:
                    temp_avg[value] = 1
                else:
                    temp_avg[value] += 1
        train_column_averages[column] = max(temp_avg, key=temp_avg.get)

# print(train_column_averages, "\n")
# print(f"Train Column Averages Length: {len(train_column_averages)}")
```

```python
test = pd.read_csv("test.csv", header=0, sep=',')
test_column_averages = {}

for column in test:
    temp_list = test[column].tolist()
    if is_numeric_dtype(test[column]):
        temp_avg = 0
        for value in temp_list:
            if not np.isnan(value):
                temp_avg += value
        test_column_averages[column] = temp_avg/len(temp_list)
    else:
        temp_avg = {}
        for value in temp_list:
            if not pd.isnull(value):
                if temp_avg.get(value) == None:
                    temp_avg[value] = 1
                else:
                    temp_avg[value] += 1
        test_column_averages[column] = max(temp_avg, key=temp_avg.get)

# print(test_column_averages, "\n")
# print(f"Test Column Averages Length: {len(test_column_averages)}")
```

```python
# for key, value in train.isnull().iteritems():
for key, value in train.iteritems():
#     print("**************************************************")
#     print(f"Column: {key}, Data Type: {value.dtype}")
#     print(value)
    temp_value = []
    column_nan_flag = 0
    for row in range(len(value.tolist())):
        if pd.isna(value[row]):
            column_nan_flag=1
            temp_value.append(train_column_averages[key])
        else:
            temp_value.append(value[row])
    if column_nan_flag == 1:
        print(f"{key} Column has NaN values. They were replaced with the row average {train_column_averages[key]}")
#     print(temp_value)
    for row in range(len(temp_value)):
        train.at[row, key] = temp_value[row]
    
```

```python
# for key, value in train.isnull().iteritems():
for key, value in test.iteritems():
#     print("**************************************************")
#     print(f"Column: {key}, Data Type: {value.dtype}")
#     print(value)
    temp_value = []
    column_nan_flag = 0
    for row in range(len(value.tolist())):
        if pd.isna(value[row]):
            column_nan_flag=1
            temp_value.append(test_column_averages[key])
        else:
            temp_value.append(value[row])
#     print(temp_value)
    print(f"{key} Column has NaN values. They were replaced with the row average {test_column_averages[key]}")
    for row in range(len(temp_value)):
        test.at[row, key] = temp_value[row]
```

```python
# Part 3

# Pick one of the variables you found to have a strong correlation with SalePrice and generate a 
# LinearRegression model using these features.  Evaluate the model using the RMSE between the logarithm 
# of the predicted value and the logarithm of the observed sale price. What was the RMSE of your model 
# and which feature did you use?
```

```python
year_built_list = train['YearBuilt'].tolist()
year_built_array = np.array(year_built_list)
# print(year_built_array)

sale_price_list = train['SalePrice'].tolist()
sale_price_array = np.array(sale_price_list)
# sale_price_array = lot_area_array.reshape(len(sale_price_array), 1)
# print(sale_price_array)

year_built_list_test = test['YearBuilt'].tolist()
year_built_array_test = np.array(year_built_list_test)
# print(year_built_array_test)

new_dataframe = pd.DataFrame(data=train['YearBuilt'])
# print(new_dataframe.shape)
# print(train['SalePrice'].T.shape)
```

```python
reg1 = LinearRegression().fit(year_built_array.reshape(-1,1), sale_price_array.reshape(-1,1))

plt.scatter(year_built_array.reshape(-1,1), sale_price_array.reshape(-1,1),  color='black')
plt.plot(year_built_array.reshape(-1,1), reg1.predict(year_built_array.reshape(-1,1)), color='blue', linewidth=3)
plt.xticks(())
plt.yticks(())
plt.show()
```

```python
prediction_values1 = reg1.predict(year_built_array_test.reshape(-1,1))
true_values1 = sale_price_array.reshape(-1,1)

# print(prediction_values1)
# print(true_values1)

for item in range(len(prediction_values1)):
    prediction_values1[item] = math.log(prediction_values1[item])
    true_values1[item] = math.log(true_values1[item])
    
# print(prediction_values1)
# print(true_values1)
```

```python
RMSE1 = math.sqrt(mean_squared_error(true_values1, prediction_values1))
print(f"I used the YearBuilt feature. Its RMSE is: {RMSE1}")
```
